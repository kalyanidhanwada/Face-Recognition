```
#Face Detection
# Cascade Classifier
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier("haarcascade_eye_tree_eyeglasses.xml")
smile_cascade = cv2.CascadeClassifier("haarcascade_smile.xml")
```

```
#Reading the image.
img = cv2.imread("vishu1.jpeg")
```

```
#Read the image as grey scale.
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```
#Select the coordinates.
faces = face_cascade.detectMultiScale(gray_img, 1.05, 5)

print(type(faces))
print(faces)
```

```
#Adding a rectangular box to detect the face.
for x, y, w, h in faces:
  img = cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0),3)
  roi_gray_img = gray_img[y:y+h, x:x+w]
  roi_color = img[y:y+h, x:x+w]
  eyes = eye_cascade.detectMultiScale(roi_gray_img)
  for (ex,ey,ew,eh) in eyes:
    cv2.rectangle(roi_color,(ex,ey),(ex+ew, ey+eh), (0,255,0),5)

resized = cv2.resize(img, (int(img.shape[1]/2),int(img.shape[0]/2)))

cv2_imshow(resized)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
